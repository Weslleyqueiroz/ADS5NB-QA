```python
import requests
import json
from IPython.display import display, HTML, Markdown, Image
```

```python
url = "https://pokeapi.co/api/v2/pokemon/pikachu"
response = requests.get(url)
```

```python
print("Status code:", response.status_code)
print("Tipo de conteúdo:", response.headers.get("content-type"))
```

```python
dados = response.json()

print(json.dumps(dados, indent=2, ensure_ascii=False)[:3000])
```

```python
nome = dados["name"]
pokemon_id = dados["id"]
altura = dados["height"]
peso = dados["weight"]
tipos = [item["type"]["name"] for item in dados["types"]]
imagem = dados["sprites"]["front_default"]

print("Nome:", nome)
print("ID:", pokemon_id)
print("Altura:", altura)
print("Peso:", peso)
print("Tipos:", tipos)
print("Imagem:", imagem)
```

```python
html_card = f'''
<div style="max-width:500px; margin:20px auto; padding:20px; border-radius:12px;
            border:1px solid #ddd; font-family:Arial; background:blue;">
    <h2 style="margin-top:0;">Pokémon encontrado: {nome.title()}</h2>
    <p><b>ID:</b> {pokemon_id}</p>
    <p><b>Altura:</b> {altura}</p>
    <p><b>Peso:</b> {peso}</p>
    <p><b>Tipos:</b> {", ".join(tipos)}</p>
    <p><b>URL da imagem:</b> {imagem}</p>
</div>
'''

display(HTML(html_card))
display(Image(url=imagem))
```

```python
assert response.status_code == 200, "Erro: a API não retornou status 200"
assert nome == "pikachu", "Erro: o nome retornado não é pikachu"
assert imagem is not None, "Erro: a imagem não foi retornada"
assert pokemon_id > 0, "Erro: o ID do Pokémon é inválido"

print("Teste executado com sucesso.")
```

```python
def buscar_pokemon(nome_pokemon):
    url = f"https://pokeapi.co/api/v2/pokemon/{nome_pokemon.lower()}"
    response = requests.get(url)
    return response
```

```python
response_charizard = buscar_pokemon("charizard")
dados_charizard = response_charizard.json()

print("Status:", response_charizard.status_code)
print("Nome:", dados_charizard["name"])
print("Imagem:", dados_charizard["sprites"]["front_default"])
display(Image(url=dados_charizard["sprites"]["front_default"]))
```

```python
response_Squirtle = buscar_pokemon("Squirtle")
dados_Squirtle = response_Squirtle.json()

print("Status:", response_Squirtle.status_code)
print("Nome:", dados_Squirtle["name"])
print("Imagem:", dados_Squirtle["sprites"]["front_default"])
display(Image(url=dados_Squirtle["sprites"]["front_default"]))
```

```python
response_erro = requests.get("https://pokeapi.co/api/v2/pokemon/pokemon_que_nao_existe")

print("Status code:", response_erro.status_code)
print("Resposta:", response_erro.text[:500])

assert response_erro.status_code == 404, "Erro: o status esperado para recurso inexistente era 404"

print("Teste negativo executado com sucesso.")
```
